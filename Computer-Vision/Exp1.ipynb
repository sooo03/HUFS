```python
# 이미지 읽기와 화면 표시

import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread("china.jpg")
plt.imshow(img)
plt.show
```

```python
img[:,100,:] = [0,255,0]
plt.imshow(img)
plt.show()
```

```python
img[100,:,:] = [0,255,255]
plt.imshow(img)
plt.show()
```

```python
r, c = img.shape[:2]
new_img = cv.resize(img, (2*r, 2*c))
plt.imshow(new_img)
plt.plot()#
```

```python
# 영상 영역의 이동
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread("china.jpg")
r, c = img.shape[:2]
M = np.float32([[1,0,100],[0,1,100]])
new_img = cv.warpAffine(img, M, (c,r))
plt.imshow(new_img)
plt.plot()
```

```python
# 영상의 회전
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread("china.jpg")
r, c = img.shape[:2]
M = cv.getRotationMatrix2D((c/2,r/2), 60, 1)
new_img = cv.warpAffine(img, M, (c,r))

plt.imshow(new_img)
plt.plot()
```

### Color interpolation 함수

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

def color_rect():
  g = np.zeros((256,256,3), np.uint8)
  for i in range(256):
    g[i,:,:] = (0, i, 255-i)

  return g

g = color_rect()

col = cv.cvtColor(g, cv.COLOR_BGR2RGB)
plt.imshow(col)
plt.show()
```
