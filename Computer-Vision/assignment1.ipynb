### 과제 1-2

Bilinear(양선형) interpolation

color_rect 함수를 수정하여 256x256 크기의 color 영상을 bilinear interpolation에 의해 생성하는 프로그램을 작성

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def bilinear_interpolation(tl, tr, bl, br, width=256, height=256):
    image = np.zeros((height, width, 3), dtype=np.uint8)

    for y in range(height):
        for x in range(width):
            weight_x = float(x) / (width - 1)
            weight_y = float(y) / (height - 1)

            top_interp = (1 - weight_x) * tl + weight_x * tr
            bottom_interp = (1 - weight_x) * bl + weight_x * br

            image[y, x] = (1 - weight_y) * top_interp + weight_y * bottom_interp

    return image
```

```python
# 네 모서리의 색상
tl = np.array([255, 0, 0])
tr = np.array([0, 0, 255])
bl = np.array([0, 255, 0])
br = np.array([255, 255, 0])

image = bilinear_interpolation(tl, tr, bl, br)

plt.imshow(image)
plt.show()
```
